```python
#Add a column of dept in 2020-2022 
data['Total Dept in 2019-2023']=data['Total Dept in 2019-2020']+data['Total Dept in 2020-2021']+data['Total Dept in 2021-2022']+data['Total Dept in 2022-2023']
data.tail()
```

```python
#remove the column of  total dept in 2019-2020,2020-2021,2021-2022,2022-2023
data.drop(["Total Dept in 2019-2020","Total Dept in 2020-2021","Total Dept in 2021-2022","Total Dept in 2022-2023"],inplace=True,axis=1)
```

```python
data
```

```python
statewise = pd.pivot_table(data,values =[" % Population below poverty line 2023", "Average Literacy Rate % 2023", "Total Population 2023 lakhs/ crores"],index="State",aggfunc=max)
```

```python
statewise["population of poverty"] = statewise[" % Population below poverty line 2023"] * statewise["Total Population 2023 lakhs/ crores"].astype(float) / 100
```

```python
statewise["population of litracy"] = statewise["Average Literacy Rate % 2023"] * statewise["Total Population 2023 lakhs/ crores"].astype(float) / 100
```

```python
statewise =statewise.sort_values(by="Total Population 2023 lakhs/ crores", ascending=False)
```

```python
statewise.style.background_gradient(cmap="Blues")
```

```python
top_10_Total_Population_2023_lakhs_crores = data.groupby(by='State').max()[['Total Population 2023 lakhs/ crores']].sort_values(by=['Total Population 2023 lakhs/ crores'], ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most population states in india",size=25)
```

```python
ax=sns.barplot(data=top_10_Total_Population_2023_lakhs_crores.iloc[:10],y="Total Population 2023 lakhs/ crores",x="State",linewidth=2,edgecolor='red')
```

```python
#top state population
top_10_Total_Population_2023_lakhs_crores = data.groupby(by='State').max()[['Total Population 2023 lakhs/ crores']].sort_values(by=['Total Population 2023 lakhs/ crores'], ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 states with most population states in india",size=25)
ax=sns.barplot(data=top_10_Total_Population_2023_lakhs_crores.iloc[:10],y="Total Population 2023 lakhs/ crores",x="State",linewidth=2,edgecolor='red')
plt.xlabel("States")
plt.ylabel("Total Population 2023 lakhs/ crores")
plt.show()
```

```python
#top state poverty rate
top_10_Total_Population_2023_lakhs_crores = data.groupby(by='State').max()[[' % Population below poverty line 2023']].sort_values(by=[' % Population below poverty line 2023'], ascending=False).reset_index()
fig=plt.figure(figsize=(12,6))
plt.title("Top 10 states with most poverty rate in india",size=25)
ax=sns.barplot(data=top_10_Total_Population_2023_lakhs_crores.iloc[:10],y=" % Population below poverty line 2023",x="State",linewidth=2,edgecolor='red')
plt.xlabel("States")
plt.ylabel("% Population below poverty line 2023")
plt.show()
```

```python
#Average Literacy Rate % 2023
top_10_Total_Population_2023_lakhs_crores = data.groupby(by='State').max()[['Average Literacy Rate % 2023']].sort_values(by=['Average Literacy Rate % 2023'], ascending=False).reset_index()
fig=plt.figure(figsize=(14,7))
plt.title("Top 10 states with most Average Literacy Rate in india",size=25)
ax=sns.barplot(data=top_10_Total_Population_2023_lakhs_crores.iloc[:10],y="Average Literacy Rate % 2023",x="State",linewidth=2,edgecolor='red')
plt.xlabel("States")
plt.ylabel("Average Literacy Rate % 2023")
plt.show()
```

```python
#Average Dept release in 2024
top_10_Total_Population_2023_lakhs_crores = data.groupby(by='State').max()[['Dept release in 2024']].sort_values(by=['Dept release in 2024'], ascending=False).reset_index()
fig=plt.figure(figsize=(14,7))
plt.title("Top 10 states with most Dept release in 2024 from Central government of india",size=25)
ax=sns.barplot(data=top_10_Total_Population_2023_lakhs_crores.iloc[:10],y="Dept release in 2024",x="State",linewidth=2,edgecolor='red')
plt.xlabel("States")
plt.ylabel("Dept release in 2024 in crore")
plt.show()
```

```python
# Debt Till 2023/Crores
top_10_Total_Population_2023_lakhs_crores = data.groupby(by='State').max()[['Debt Till 2023/Crores']].sort_values(by=['Debt Till 2023/Crores'], ascending=False).reset_index()
fig=plt.figure(figsize=(14,7))
plt.title("Top 10 states with most Dept till 2023 ",size=25)
ax=sns.barplot(data=top_10_Total_Population_2023_lakhs_crores.iloc[:10],y="Debt Till 2023/Crores",x="State",linewidth=2,edgecolor='red')
plt.xlabel("States")
plt.ylabel("Debt Till 2023/Crores")
plt.show()
```
